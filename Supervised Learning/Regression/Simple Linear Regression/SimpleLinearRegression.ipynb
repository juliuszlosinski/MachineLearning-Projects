### 1. Importing libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

### 2. Importing dataset

```python
data_frame = pd.read_csv("Salary_Data.csv")
X = data_frame.iloc[:, :-1].values
y = data_frame.iloc[:, -1].values

print(f"X: {X}, \ny: {y}")
```

### 3. Splitting data into training and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3)

print(f"X_train: \n{X_train}, \ny_train: {y_train}")
print(f"X_test: \n{X_test}, \ny_test: {y_test}")
```

### 4. Training Simple Linear Regression on the training set

```python
model = LinearRegression()
model.fit(X_train, y_train)

print(f"Coefficient: {model.coef_[0]}")
print(f"Mean squared error: {mean_squared_error(y_test, model.predict(X_test))}")
print(f"Coeffiecient of determination: {r2_score(y_test, model.predict(X_test))}")
```

### 5. Predicting testing results

```python
y_predicted = model.predict(X_test)

print(f"y ground truth: {y_test}")
print(f"y predicted: {y_predicted}")
```

### 6. Visualising the training results

```python
plt.scatter(X_train, y_train, color="green")
plt.plot(X_train, model.predict(X_train), color="blue")
plt.title("Salary(experience)")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.grid(True)
plt.plot()
```

### 7. Visualising test results

```python
plt.scatter(X_test, y_test, color="red")
plt.plot(X_test, model.predict(X_test), color="blue")
plt.title("Salary(experience)")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

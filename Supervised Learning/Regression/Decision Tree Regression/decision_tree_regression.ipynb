### Decision Tree Regression

<b style="color:green">GOAL:</b> Finding the optimal split.

Final leaves = Terminal leaves




### 1. Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
data_frame = pd.read_csv("./Position_Salaries.csv")

X = data_frame.iloc[:, 1:-1].values
y = data_frame.iloc[:, -1].values

print(f"Data frame: \n{data_frame}")
print(f"X: \n{X}")
print(f"y: \n{y}")
```

### 2. Fitting Decision Tree Regression model.

```python
decision_tree_regression_model = DecisionTreeRegressor(random_state=0)
decision_tree_regression_model.fit(X, y)
```

### 3. Testing Decision Tree Regression model.

```python
y_predicted = decision_tree_regression_model.predict(
    [
        [5],
        [6],
        [7],
        [10],
        [11]
    ]
)
print(f"y_predicted: \n{y_predicted}")

y_predicted = decision_tree_regression_model.predict(
    np.array([5, 6, 7, 10, 11]).reshape(-1, 1)
)

print(f"y_predicted: \n{y_predicted}")

y_predicted = decision_tree_regression_model.predict(X)

print(f"\nX: \n{X}")
print(f"y_predicted: \n{y_predicted}")
```

### 4. Visualising the Decision Tree Regression model.

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color="red")
plt.plot(X_grid, decision_tree_regression_model.predict(X_grid), color="blue")
plt.title("Salary(Level) - Decision Tree Regression Model")
plt.xlabel("Level [-]")
plt.ylabel("Salary [$]")
plt.grid(True)
plt.savefig("Salary(Level)-Decision_Tree_Regression.jpg")
```

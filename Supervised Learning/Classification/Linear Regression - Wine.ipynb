<h3>0. Importing librariers</h3>

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

<h3>1. Loading wine data</h3>

```python
data_set = load_wine()
X_data = data_set.data[0:100, 0:2]   # alcohol | malic_aid
y_data = data_set.data[0:100, 3]     # ash

print(f"ALCOHOL | MALIC_AID")
print(f"X: \n{X_data}")
print(f"ASH")
print(f"y: {y_data}")
```

<h3>2. Splitting data into training and testing data set.</h3>

```python
X_training_data, X_testing_data, y_training_data, y_testing_data = train_test_split(X_data, y_data, train_size=0.5)

print(f"Dimension of X training data: {X_training_data.shape}")
print(f"Dimension of y training data: {y_training_data.shape}")
print(f"Dimension of X testing data: {X_testing_data.shape}")
print(f"Dimension of y testing data: {y_testing_data.shape}")
```

<h3>3. Standardizing data:</h3>



```python
standard_scaler = StandardScaler()
X_training_standarded_data = standard_scaler.fit_transform(X_training_data)
X_testing_standarded_data = standard_scaler.fit_transform(X_testing_data)

print(f"Standarded X training data: {X_training_standarded_data}")
print(f"Standarded X testing data: {X_testing_standarded_data}")
```

<h3>4. Implementing and learning <b>LinearRegression Model</b><h3>



```python
linear_regression_model = LinearRegression()
linear_regression_model.fit(X_training_standarded_data, y_training_data)
```

<h3>5. Testing learned LinearRegression Model</h3>

```python
y_predicted_data = linear_regression_model.predict(X_testing_standarded_data)

print(f"X_testing_standarded_data: {X_testing_standarded_data}")
print(f"y_testing_data: {y_testing_data}")
print(f"y_predicted: {y_predicted_data}")

mean_absolute_error_result = mean_absolute_error(y_testing_data, y_predicted_data)
mean_squared_error_result = mean_squared_error(y_testing_data, y_predicted_data)
r2_result = r2_score(y_testing_data, y_predicted_data)

print("The LinearRegression Model performance for testing set:")
print("--------------------------------------------------------")
print(f"Mean absolute error: {mean_absolute_error_result}")
print(f"Mean squared error: {mean_squared_error_result}")
print(f"R2 score: {r2_result}")
```

<h3>6. Drawing/ displaying results on 3D Scatter</h3>

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.set_xlabel("Alcohol")
ax.set_ylabel("Malic aid")
ax.set_title("Ash(alcohol, Malic Aid)")
ax.scatter3D(X_testing_data[0:20, 1], X_testing_data[0:20, 1], y_testing_data[0:20])
ax.scatter3D(X_testing_data[0:20, 1], X_testing_data[0:20, 1], y_predicted_data[0:20])
plt.show()
```

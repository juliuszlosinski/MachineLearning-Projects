### <b>Hierachical Clustering (HC) - Dendograms</b>

### <b> Data preprocessing</b>

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dataset = pd.read_csv("Mall_Customers.csv")

print(f"Dataset: \n{dataset}\n")

description = dataset.describe()

print(f"Description: \n{description}\n")

X = dataset.iloc[:, [3,4]].values

print(f"X: \n{X}\n")
```

### <b>Finding the best/optimal number of clusters by using dendogram</b>

```python
dendogram = sch.dendrogram(sch.linkage(X, method = "ward"))
plt.xlabel("Customers")
plt.ylabel("Euclidean distances")
plt.title("Dendogram")
plt.savefig("Dendogram-Hierarchical_Clustering_agglomerative_ward.jpg")
plt.show()
```

### <b>Training the Hierarchical Clustering (HC) model on the dataset (getting dependent variable)</b>

```python
hierarchical_clustering_model = AgglomerativeClustering(n_clusters = 5, affinity = "euclidean", linkage = "ward")
y = hierarchical_clustering_model.fit_predict(X)
print(f"y: \n{y}\n")
```

### <b>Visualizing the clusters</b>

```python
plt.scatter(X[y==0, 0], X[y==0, 1], color="blue", label="Cluster 1")
plt.scatter(X[y==1, 0], X[y==1, 1], color="red", label="Cluster 2")
plt.scatter(X[y==2, 0], X[y==2, 1], color="purple", label="Cluster 3")
plt.scatter(X[y==3, 0], X[y==3, 1], color="green", label="Cluster 4")
plt.scatter(X[y==4, 0], X[y==4, 1], color="orange", label="Cluster 5")
plt.grid(True)
plt.xlabel("Average income (0-140) [k$]")
plt.ylabel("Average spending (1-100) [-]")
plt.title("Clusters of customers - Hierarchical Clustering (HC)")
plt.savefig("hierarchical_clustering.jpg")
plt.show()
```
